```python
url = 'https://www.basketball-reference.com/leagues/NBA_2019_per_game.html'
```

```python
import pandas as pd
```

```python
df=pd.read_html(url)
df
```

```python
len(df)
```

```python
df[0]
```

```python
df2019=df[0]
```

```python
df2019[df2019.Age=='Age']
```

```python
len(df2019[df2019.Age=='Age'])
```

```python
df = df2019.drop(df2019[df2019.Age == 'Age'].index)
```

```python
import seaborn as sns
```

```python
sns.distplot(df.PTS, 
             kde=False)
```

```python
sns.distplot(df.PTS, 
             kde=False,
             hist_kws=dict(edgecolor="black", linewidth=2))
```

```python
sns.distplot(df.PTS,
             kde=False,
             hist_kws=dict(edgecolor="black", linewidth=2),
             color='#00BFC4')
```
